```python
# Standard Libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Functions
def make_table(data, figsize=(12,8)):
    # Make fig
    fig, ax = plt.subplots(figsize=figsize)

    # Hide axes
    fig.patch.set_visible(False)
    ax.axis('off')
    ax.axis('tight')
    
    # Make column headings
    columns = []
    for c in data.columns:
        columns.append(r'{}'.format(c))

    # Make table
    ax.table(cellText=data.values, colLabels=columns, loc='center', cellLoc='center', colWidths=[1/13, 4/13, 4/13, 2/13, 2/13])

    # Fig tight layout
    fig.tight_layout()
    
    return fig, ax

def plot_table(ax, save = False, filename = 'default.pdf'):
    if save:
        plt.savefig(filename)
    plt.show()
```

```python
# Load csv fra en ydre mappe '..' går en mappe tilbage og '/Data' går ind i den ydre mappe Data
df = pd.read_csv('../Data/NIST.csv')
```

```python
# Visualiser dataframe
df.head(20)
```

```python
# Make figure
fig, ax = make_table(df, figsize=(10,8))

# Plot figure
plot_table(ax, save=True, filename=f'../../../Final_graphs/nist_data_table.pdf')
```
