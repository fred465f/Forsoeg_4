```python
# Standard Libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Functions
def make_table(data, figsize=(12,8)):
    # Make fig
    fig, ax = plt.subplots(figsize=figsize)

    # Hide axes
    fig.patch.set_visible(False)
    ax.axis('off')
    ax.axis('tight')
    
    # Make column headings
    columns = []
    cellText = []
    for i, c in enumerate(data.columns):
        if i <= 2:
            columns.append(r'{}'.format(c))
        if i > 4:
            columns.append(r'{}'.format(c))
    for row in data.to_numpy():
        cellText.append([r'{}'.format(row[i]) for i in range(7) if not i in [3,4]])
    
    # Make table
    table = ax.table(cellText=cellText, colLabels=columns, loc='center', cellLoc='center', 
             colWidths=[1/15, 3/15, 3/15, 4/15, 4/15])
    
    # Set fontsize
    table.set_fontsize(26)
    table.scale(1.5, 2.0)

    # Fig tight layout
    fig.tight_layout()
    
    return fig, table

def plot_table(ax, save = False, filename = 'default.pdf'):
    if save:
        plt.savefig(filename)
    plt.show()
```

```python
# Load csv fra en ydre mappe '..' går en mappe tilbage og '/Data' går ind i den ydre mappe Data
df = pd.read_csv('../Data/NIST.csv')
df.replace(np.NaN, value='', inplace=True)
```

```python
# Visualiser dataframe
df.head(20)
```

```python
# Make figure
fig, ax = make_table(df, figsize=(12,16))

# Plot figure
plot_table(ax, save=True, filename=f'../../../Final_graphs/nist_data_table.pdf')
```
